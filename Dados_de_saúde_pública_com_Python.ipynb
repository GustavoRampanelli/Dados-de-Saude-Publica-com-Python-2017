Importando Biblioteca

```python
import pandas as pd

pd.options.display.max_columns = None
data = pd.read_csv('/content/1.1 ANIMBR17.csv', encoding='latin1')

data
```

```python
data.columns
```

```python
keep_columns = ['DT_NOTIFIC', 'SEM_NOT', 'NU_ANO', 'SG_UF_NOT',
       'ID_MUNICIP', 'ID_REGIONA', 'DT_SIN_PRI', 'SEM_PRI', 'DT_NASC',
       'NU_IDADE_N', 'CS_SEXO', 'CS_GESTANT', 'CS_RACA', 'CS_ESCOL_N', 'SG_UF',
       'ID_MN_RESI', 'ID_RG_RESI', 'ID_PAIS', 'DT_INVEST', 'ID_OCUPA_N',
       'ANT_DT_ACI', 'ANT_UF', 'ANT_MUNIC_', 'ANT_LOCALI', 'ANT_ZONA',
       'ANT_TEMPO_', 'ANT_LOCA_1', 'MCLI_LOCAL', 'CLI_DOR', 'CLI_EDEMA',
       'CLI_EQUIMO', 'CLI_NECROS', 'CLI_LOCAL_', 'CLI_LOCA_1', 'MCLI_SIST',
       'CLI_NEURO', 'CLI_HEMORR', 'CLI_VAGAIS', 'CLI_MIOLIT', 'CLI_RENAL',
       'CLI_OUTR_2', 'CLI_OUTR_3', 'CLI_TEMPO_', 'TP_ACIDENT', 'ANI_TIPO_1',
       'ANI_SERPEN', 'ANI_ARANHA', 'ANI_LAGART', 'TRA_CLASSI', 'CON_SOROTE',
       'NU_AMPOLAS', 'NU_AMPOL_1', 'NU_AMPOL_8', 'NU_AMPOL_6', 'NU_AMPOL_4',
       'NU_AMPO_7', 'NU_AMPO_5', 'NU_AMPOL_9', 'NU_AMPOL_3', 'COM_LOC',
       'COM_SECUND', 'COM_NECROS', 'COM_COMPOR', 'COM_DEFICT', 'COM_APUTAC',
       'COM_SISTEM', 'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC', 'COM_CHOQUE',
       'DOENCA_TRA', 'EVOLUCAO', 'DT_OBITO', 'DT_ENCERRA', 'DT_DIGITA']
```

```python
data = data[keep_columns].reset_index(drop=True)

data
```

```python
data['MCLI_LOCAL'] = data['MCLI_LOCAL'].replace({1:'Sim', 2:'Não', 9:'Ignorado'})

data
```

```python
columns_yes_no = ['MCLI_LOCAL','CLI_DOR','CLI_EDEMA',
                  'CLI_EQUIMO','CLI_NECROS','CLI_LOCAL_',
                  'MCLI_SIST','CLI_NEURO','CLI_HEMORR',
                  'CLI_VAGAIS','CLI_MIOLIT','CLI_RENAL',
                  'CLI_OUTR_2','CON_SOROTE','COM_LOC',
                  'COM_SECUND','COM_NECROS','COM_COMPOR',
                  'COM_RENAL','COM_EDEMA','COM_SEPTIC',
                  'COM_CHOQUE','DOENCA_TRA']

for columns in columns_yes_no:
  data[columns] = data[columns].replace({1:'Sim', 2:'Não', 9:'Ignorado'})

data
```

```python
data.info ()
```

```python
columns_date = ['DT_NOTIFIC', 'DT_SIN_PRI', 'DT_NASC',
                'DT_INVEST', 'DT_OBITO', 'DT_ENCERRA',
                'DT_DIGITA', 'ANT_DT_ACI']

for columns in columns_date:
  data[columns] = pd.to_datetime(data[columns], format='%Y%m%d', errors='coerce')

data
```

```python
data['ANT_ZONA'] = data['ANT_ZONA'].replace({1:'Urbana',
                                             2:'Rural',
                                             3:'Periurbana',
                                             9:'Ignorado'})

data['ANT_TEMPO_'] = data['ANT_TEMPO_'].replace({  1: '0 - 1h',
                                                    2: '1 - 3h',
                                                    3: '3 - 6h',
                                                    4: '6 - 12h',
                                                    5:'12 - 24h',
                                                    6:'mais de 24h',
                                                    9: 'Ignorado'})

data['ANT_LOCA_1'] = data['ANT_LOCA_1'].replace({ 1 : 'Cabeça',
                                                    2 : 'Braço',
                                                    3 : 'Ante-Braço',
                                                    4 : 'Mão',
                                                    5 : 'Dedo da Mão',
                                                    6 : 'Tronco',
                                                    7 : 'Coxa',
                                                    8 : 'Perna',
                                                    9 : 'Pé',
                                                    0 : 'Dedo do Pé',
                                                    99 : 'Ignorado' })

data['CLI_TEMPO_'] = data['CLI_TEMPO_'].replace({ 1 : 'Normal',
                                                    2 : 'Alterado',
                                                    9 : 'Não realizado'})

data['TP_ACIDENT'] = data['TP_ACIDENT'].replace({ 1 : 'Serpente',
                                                    2 : 'Aranha',
                                                    3 : 'Escorpião',
                                                    4 : 'Lagarta',
                                                    5 : 'Abelha',
                                                    6 : 'Outros',
                                                    9 : 'Ignorado' })

data['ANI_SERPEN'] = data['ANI_SERPEN'].replace({ 1 : 'Botrópico',
                                                    2 : 'Crotálico',
                                                    3 : 'Elapídico',
                                                    4 : 'Laquético',
                                                    5 : 'Serpente não peçonhenta',
                                                    9 : 'Ignorado' })

data['ANI_ARANHA'] = data['ANI_ARANHA'].replace({ 1 : 'Foneutrismo',
                                                    2 : 'Loxoscelismo',
                                                    3 : 'Latrodectismo',
                                                    4 : 'Outra aranha',
                                                    9 : 'Ignorado'})

data['ANI_LAGART'] = data['ANI_LAGART'].replace({ 1 : 'Lonomia',
                                                    2 : 'Outra lagarta',
                                                    9 : 'Ignorado'})

data['TRA_CLASSI'] = data['TRA_CLASSI'].replace({ 1 : 'Leve',
                                                    2 : 'Moderado',
                                                    3 : 'Grave',
                                                    9 : 'Ignorado' })

data['EVOLUCAO'] = data['EVOLUCAO'].replace({ 1 : 'Cura',
                                                2 : 'Óbito por acidente por animais peçonhentos',
                                                3 : 'Óbito por outras causas',
                                                9 : 'Ignorado'  })


data['CS_GESTANT'] = data['CS_GESTANT'].replace({ 1: '1º Trimestre',
                                                    2: '2º Trimestre',
                                                    3: '3º Trimestre',
                                                    4: 'Idade gestacional ignorada',
                                                    5: 'Não',
                                                    6: 'Não se aplica',
                                                    9: 'Ignorado'  })

data['CS_RACA'] = data['CS_RACA'].replace({ 1: 'Branca',
                                              2: 'Preta',
                                              3: 'Amarela',
                                              4: 'Parda',
                                              5: 'Indígena',
                                              9: 'Ignorado' })

data['CS_ESCOL_N'] = data['CS_ESCOL_N'].replace({ 0: 'Analfabeto',
                                                    1: '1ª a 4ª série incompleta do EF',
                                                    2: '4ª série completa do EF ( antigo 1° grau)',
                                                    3: '5ª à 8ª série incompleta do EF (antigo ginásio ou 1° grau)',
                                                    4: 'Ensino fundamental completo (antigo ginásio ou 1° grau)',
                                                    5: 'Ensino médio incompleto (antigo colegial ou 2° grau)',
                                                    6: 'Ensino médio completo (antigo colegial ou 2° grau)',
                                                    7: 'Educação superior incompleta',
                                                    8: 'Educação superior completa',
                                                    9: 'Ignorado',
                                                    10: 'Não se aplica' })

data
```

```python
ages = []

for i in data['NU_IDADE_N']:
  if i > 4000:
    ages.append(i - 4000)

  elif 4000 > i:
    ages.append(0)

  else:
    ages.append(0)

data['NU_IDADE_N'] = ages

ages
```

Codigo do IBGE

```python
municipios = pd.read_csv('https://raw.githubusercontent.com/andrejarenkow/Curso_SINAN_Udemy/main/dados/populacao%20ibge%206%20municipio%20br.csv')
municipios = municipios[['IBGE6', 'Municipio']]
municipios = municipios.set_index('IBGE6')

dicionario_municipios = municipios.to_dict()['Municipio']

data['ID_MUNICIP_nome'] = data['ID_MUNICIP'].map(dicionario_municipios)
data['ID_MN_RESI_nome'] = data['ID_MN_RESI'].map(dicionario_municipios)
data['ANT_MUNIC_nome'] = data['ANT_MUNIC_'].map(dicionario_municipios)

data
```

Ocupação

```python
ocupacao = pd.read_csv('https://raw.githubusercontent.com/andrejarenkow/Curso_SINAN_Udemy/main/dados/CBO2002%20-%20Ocupacao.csv')
ocupacao = ocupacao.set_index('CODIGO')
dicionario_ocupacao = ocupacao.to_dict()['TITULO']
data['ID_OCUPA_N_nome'] = data['ID_OCUPA_N'].map(dicionario_ocupacao)
data
```

Unidade Federativa

```python
tabela_uf = pd.read_csv('https://raw.githubusercontent.com/andrejarenkow/Curso_SINAN_Udemy/main/dados/cod_uf.csv')
tabela_uf = tabela_uf.set_index('Código UF')
dicionario_uf = tabela_uf.to_dict()['UF']
data['SG_UF_nome'] = data['SG_UF'].map(dicionario_uf)
data['SG_UF_NOT_nome'] = data['SG_UF_NOT'].map(dicionario_uf)
data['ANT_UF_nome'] = data['ANT_UF'].map(dicionario_uf)
data
```
